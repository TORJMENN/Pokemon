# Selection

#### How many Fire-Flying Pokemons are there? (5)

```python
df.loc[(df['Type 1']=='Fire')&(df['Type 2']=='Flying'),'Type 2'].shape
```

#### How many 'Poison' pokemons are across both types? (59)

```python
df.loc[(df['Type 1']=='Poison')|(df['Type 2']=='Poison'),].shape
```

#### What pokemon of Type 1 Ice has the strongest defense? (Avalugg)

```python
df.loc[df['Type 1'] == 'Ice'].sort_values(by='Defense', ascending=False).head()
```

#### What's the most common type of Legendary Pokemons? (Psychic)

```python
df.loc[df['Legendary'] == True,'Type 1'].value_counts().head(1)
```

#### the most powerful Dragon from the last two generations

```python
df.loc[(df['Generation']>=5)&(df['Type 1']=='Water')].sort_values(by='Total',ascending=False).head(1)
```

#### Select most powerful Fire-type pokemons

```python
powerful_fire_df = df.loc[(df['Attack']>100)&(df['Type 1']=='Fire')]
powerful_fire_df
```

#### Select all Water-type, Flying-type pokemons

```python
water_flying_df = df.loc[(df['Type 1']=='Water')&(df['Type 2']=='Flying')]
water_flying_df
```

#### Select specific columns of Legendary pokemons of type Fire

```python
legendary_fire_df = df.loc[(df['Type 1']=='Fire')&(df['Legendary']==True),['Name','Attack','Generation']]
legendary_fire_df
```

#### Select Slow and Fast pokemons

```python
ax = df['Speed'].plot(kind='hist', figsize=(10, 5), bins=100)
ax.axvline(df['Speed'].quantile(.05), color='red')
ax.axvline(df['Speed'].quantile(.95), color='red')
```

```python
slow_fast_df = df.loc[
    (df['Speed'] < df['Speed'].quantile(.05)) |
    (df['Speed'] > df['Speed'].quantile(.95))
]
slow_fast_df
```

#### the Ultra Powerful Legendary Pokemon (Rayquaza)

```python
fig, ax = plt.subplots(figsize=(14, 7))
sns.scatterplot(data=df, x="Defense", y="Attack", hue='Legendary', ax=ax)
ax.annotate(
    "Who's this guy?", xy=(140, 150), xytext=(160, 150), color='red',
    arrowprops=dict(arrowstyle="->", color='red')
)
```
